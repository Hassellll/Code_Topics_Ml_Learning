```python
#We import the libraries that we will be using
import pandas as pd
from scipy import stats
```

```python
#We have it read the CSV file to process it and work with it.
datos = pd.read_csv("T_test_Anova.csv")
```

```python
#We define the variables separately
A = datos["model_A"]
B = datos["model_B"]
C = datos["model_C"]
```

```python
# 3. T-TEST (A vs B)
#Compare two models
# -------------------------
t, p = stats.ttest_rel(A, B)

print("T-test (A vs B)")
print("p-value:", p)

if p < 0.05:
    print("At least one model is different")
else:
    print("There is no significant difference.")
```

```python
# 4. ANOVA (A vs B vs C)
#Compare three or more models
# -------------------------
f, p = stats.f_oneway(A, B, C)

print("ANOVA (A vs B vs C)")
print("p-value:", p)

if p < 0.05:
    print("At least one model is different")
else:
    print("There is no significant difference.")
```

```python
#Important rule to remember:
#p < 0.05 → there is a difference
#p ≥ 0.05 → there is no difference
```
