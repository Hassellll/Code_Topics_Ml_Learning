```python
#Separar X (características) e y (objetivo)
X = datos[["Edad","Salario"]]
Y = datos["Compro"]
```

```python
#We converted to Numoy
X = X.values
```

```python
# # Add column of 1s (bias that is ones)
ones = np.ones((X.shape[0], 1))
design_matrix = np.hstack((ones, X))

print(design_matrix)
```
