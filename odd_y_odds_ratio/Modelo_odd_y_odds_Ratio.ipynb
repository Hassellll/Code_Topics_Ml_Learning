What does this code do?

This code uses a logistic regression model to analyze how certain variables (age, weight, and smoking status) influence the likelihood of a person developing a disease.


```python
#We import the libraries

import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

```

```python
#We read the data 
datos = pd.read_csv("datos_salud.csv")
```

```python
#We use columns to avoid mistakes when writing variables
datos.columns
```

```python
#We define X and Y by placing the variables
X = datos[["Edad","Peso","Fumador"]]
Y = datos["Enfermedad"]
```

```python
#we create the model and we train it
Modelo = LogisticRegression()
Modelo.fit(X,Y)
```

```python
#Coefficients
coeficientes = Modelo.coef_[0]
variables = X.columns
```

```python
print("==Coefficients==")
for var,coef in zip(variables,coeficientes):
    print(f"{var}: {coef:4f}")
```

```python

#We calculate Odds Ratio
#Odds Ratio is the comparison between two groups

#Odds ratio measures how a one-unit increase in a variable
#affects the odds of the outcome


odds_ratio = np.exp(coeficientes)
```

```python
print("==ODDS RATIO==")
for var,or_val in zip(variables,odds_ratio):
    print(f"{var}: {or_val:.4f}")
```

```python
#Probabilidades predichas
probabilidades = Modelo.predict_proba(X)[:,1]
```

```python
#We calculate Odds:

#Ratio means how likely something is to happen and how likely it is that something will not happen
#Odds = 1 → 50% probability
#Odds > 1 → more likely to occur
#Odds < 1 → less likely
odds = probabilidades / (1-probabilidades)
```

```python
print("\n==Odds Example==")
for i in range(5):
    print(f"Persona {i+1}: Prob={probabilidades[i]:.4f}-> odds={odds[i]:.4f}")
```

```python
#What the data want to tell us ?

#The data already collected gives us valuable information, since if the odds are greater than 5 there is a higher probability that people will get sick, 
# and conversely,if they are less than 5 there is a lower probability that these people will get sick.
```
