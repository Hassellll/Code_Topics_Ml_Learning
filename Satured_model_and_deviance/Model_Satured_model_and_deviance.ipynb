```python
#We import libraries
import pandas as pd
import statsmodels.api as sm
```

```python
#Load the dataset
data = pd.read_csv("Satured_model_and_deviance.csv")
```

```python
#Preview our data
data.head()
```

```python
#data information
data.info()
```

```python
#Statistical summary
data.describe()
```

```python
#We use columns to do not make mistakes defining 

data.columns
```

```python
#Feature selection
X = data[["Age","Income"]]
X = sm.add_constant(X) #add Intercept
Y = data["Buy"]
```

```python
#Logistic regression model (GLM)
model = sm.GLM(Y,X,family=sm.families.Binomial())
result =model.fit()
```

```python
#Model summary
print(result.summary())
```

```python
## Deviance analysis
print(f"Deviance: {result.deviance}")
print(f"Null Deviance: {result.null_deviance}")
```

#Interpretation
#The model deviance is lower than the null deviance, indicating that
#the model provides a better fit than a model with no predictors.
