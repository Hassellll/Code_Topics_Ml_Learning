```python
#Importamos las librerias que estaremos usando 

import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
#Dejamos que pandas lea el archivo csv y lo convierta

datos = pd.read_csv("DiabetesSioNo.csv")
```

```python
#Observamos los datos en tabla

datos.head()
```

```python
datos.info()
```

```python
datos.describe() 
```

```python
#Es recomendable usar columns para no confundirse escribiendo mal el nombre de las variables

datos.columns
```

```python
#Definimos X y Y colocandoles las variables 

x = datos[["Edad","IMC","Glucosa","Presion","Insulina","Actividad_Fisica"]]
y = datos["Diabetes"]
```

```python
#como los tenemos ya definidos en este caso primero usaremos la estructura de fit :
#FIT :
#Entrena y prueba una sola vez
#Tiene mas riesgo de overfitting 
#Entrena con los datos que tiene 
#Es mas rapido 
#No recomendable usar con pocos datos

#1 separamos el modelo con train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)



```

```python
#2 Creamos el modelo y lo entrenamos
modelo = LogisticRegression(max_iter=1000)
modelo.fit(x_train,y_train)
```

```python
#3 Lo evaluamos 
pre = modelo.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,pre)
print(accuracy)
```

```python
#El accuracy indica el porcentaje de predicciones correctas del modelo.
#Por ejemplo, 0.66 significa que el modelo acierta aproximadamente el 66% de las veces.
```

```python
datos.head()
```

```python
#Hacemos una prediccion de prueba

nueva_prediccion = [[28,23.2,70,90,2,0]]
prediccion= modelo.predict(nueva_prediccion)

if prediccion[0] == 1:
    print("Tienes Diabetes")
else:
    print("No tienes Diabetes")
```

```python
#Ahora usaremos Cross_val_score

cross = cross_val_score(modelo,x,y, cv=5)
print(cross)
```

```python
#El modelo se entrena 5 veces usando diferentes divisiones de los datos.
#Cada valor representa el accuracy obtenido en cada fold.
#Podemos calcular el promedio para obtener el rendimiento general.
```

```python
#Verificamos el rednimiento del modelo:

modelo.score(x,y)
```
